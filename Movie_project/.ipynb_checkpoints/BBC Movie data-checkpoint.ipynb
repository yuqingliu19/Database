```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('movies_list.csv')
df
```

```python
df.groupby('critic_cn')['critic_name'].nunique()
```

```python
df1 = df.groupby('critic_cn')['movie_name'].value_counts().reset_index(name='count')
df1
```

```python
df1["string"] = df1["movie_name"] + ": " + df1["count"].map(str) + np.where(df1["count"]>1, ' votes', ' vote')
df1.head()
df1.loc[30:]
```

```python
output = df1.groupby('critic_cn')['string'].apply(lambda x: "<p>%s</p>" % '</p><p> '.join(x)).reset_index(name='article')
output
```

```python
output
```

```python
output['headline'] = output['headline'].map(str) + " critic"
```

```python
df = pd.read_csv('director_info.csv')
df
```

```python
df.rename(columns={'name': 'movie_director'}, inplace=True)
df
```

```python
output_d = pd.merge(df_m,df[['birth_place','birth_year','movie_director']],left_on='movie_director', right_on='movie_director', how='left')
```

```python
output_d
```

```python
output_d['movie_name'] = output_d['movie_name'].apply(lambda x: x.strip())
output_d.drop_duplicates(subset='movie_name',keep='first', inplace=True)
output_d
```

```python
output_d[output_d['movie_director'] == 'Ang Lee']
```

```python
output_d.dropna(subset=['birth_year'], inplace=True)
output_d
```

```python
output_d = output_d[output_d['birth_year'].str.contains('^\d\d\d\d$', na=False)]
output_d
```

```python
output_d['color'] = "#000066"
output_d
```

```python
output_d['age'] = 2017 - output_d['birth_year'].astype(int)
output_d
```

```python
all_movies = output_d.groupby('movie_director')['movie_name'].apply(lambda x: "<p>%s</p>" % '</p><p> '.join(x)).reset_index(name='all_movies')
```

```python
all_movies
```

```python
output_e = pd.merge(output_d,all_movies[['all_movies','movie_director']],left_on='movie_director', right_on='movie_director', how='left')
```

```python
output_e 
```

```python
output_e.reset_index(drop=True, inplace=False)
```

```python
directors = gpd.GeoDataFrame(output_e)
```

```python
from geopy.geocoders import Nominatim
from shapely.geometry import Point

def geocode_row(row):
    try:
        geopy_point = geolocator.geocode(row["birth_place"]).point
        return Point(geopy_point.longitude, geopy_point.latitude)
    except:
        pass
    
geolocator = Nominatim()
directors['geometry'] = directors.apply(geocode_row, axis=1)
```

```python
directors2 = directors.dropna()
```
